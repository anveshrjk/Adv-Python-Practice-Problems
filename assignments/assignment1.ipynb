```python
#1. Reverse an Array - O(n)

arr = [1, 2, 3, 4, 5]

left = 0
right = len(arr) - 1

while left < right:
    arr[left], arr[right] = arr[right], arr[left]
    left += 1
    right -= 1

print(arr)
```

```python
# 2. Find the Maximum & Minimum Element - O(n)

def Min_Max(nums):

    if not nums:
        return None

    min = max = nums[0]


    for i in nums:
        if i < min:
            min = i

        elif max < i:
            max = i
    return [min, max]

Min_Max([1,54,2,325,0])
```

```python
# 3. Sum Of Elements - O(n)

def Arr_sum(nums):

    total = 0

    for val in nums:
        total = total + val

    return total

Arr_sum([234,24,3,45,])
```

```python
# 4. Find the Second Largest Element - O(n)

def secondLargest(nums):

  if len(nums) < 2:
    return None

  largest = float('-inf')
  second = float('-inf')

  for i in nums:

    if i > largest:
      second = largest
      largest = i

    elif i > second and i != largest:
      second = i

    if second == float('-inf'):
      return None

    else:
      return second

print(secondLargest([12,4,3,7,2]))
```

```python
# 5. Count Frequency of Elements - O(n)

def countFrequencies(arr):

    freq_map = {}

    for num in arr:

        if num in freq_map:
            freq_map[num] += 1

        else:
            freq_map[num] = 1

    return freq_map

arr = [10, 20, 20, 10, 10, 30]
print(countFrequencies(arr))

'''
# or use counter:
from collections import Counter

arr = [10, 20, 20, 10, 10, 30]
freq_map = Counter(arr)

print(freq_map)
'''
```

```python
# 6. Check if Array is Sorted - O(n)

def isSorted(arr):
    if len(arr) <= 1:
        return True

    for i in range(1, len(arr)):
        if arr[i - 1] > arr[i]:
            return False

    return True

print(isSorted([1, 2, 3, 4, 5]))
print(isSorted([1, 2, 5, 3, 4]))
```

```python
# 7. Rotate Array by k Positions: Rotate the array to the right by k positions.

def rotate(nums, k):
    n = len(nums)
    if n == 0:
        return

    k = k % n

    def reverse(start, end):
        while start < end:
            nums[start], nums[end] = nums[end], nums[start]
            start += 1
            end -= 1

    reverse(0, n - 1)
    reverse(0, k - 1)
    reverse(k, n - 1)

nums1 = [1, 2, 3, 4, 5, 6, 7]
rotate(nums1, 3)
print(nums1)

nums2 = [-1, -100, 3, 99]
rotate(nums2, 2)
print(nums2)
```

```python
# 8. Find Pair with Given Sum: Find a pair of elements that adds up to a target sum.

def findPair(nums, target):
    seen = set()

    for num in nums:
        comp = target - num

        if comp in seen:
            return [comp, num]

        seen.add(num)

    return []

print(findPair([8, 7, 2, 5, 3, 1], 10))
print(findPair([5, 2, 6, 8, 1, 9], 12))
print(findPair([1, 2, 3, 4], 10))
```

```python
# 9. Remove Duplicates from Array: Remove duplicates from the array while maintaining order.

def removeDuplicates(nums):
    seen = set()
    result = []

    for num in nums:
        if num not in seen:
            seen.add(num)
            result.append(num)

    return result

print(removeDuplicates([1, 2, 3, 2, 1, 4]))
print(removeDuplicates([5, 5, 5, 5]))

'''
def removeDuplicates(nums):
    return list(dict.fromkeys(nums))
'''
```

```python
# 10. Merge Two Sorted Arrays - O(n+m)

def mergeSorted(nums1, nums2):
    result = []
    i = 0
    j = 0

    while i < len(nums1) and j < len(nums2):
        if nums1[i] < nums2[j]:
            result.append(nums1[i])
            i += 1
        else:
            result.append(nums2[j])
            j += 1

    while i < len(nums1):
        result.append(nums1[i])
        i += 1

    while j < len(nums2):
        result.append(nums2[j])
        j += 1

    return result

print(mergeSorted([1, 3, 5], [2, 4, 6]))
print(mergeSorted([10, 20], [5, 15, 25, 35]))
print(mergeSorted([], [1, 2]))
```

```python
# 11. Remove given Element from Array - O(n)

def removeElement(nums, val):
    writer = 0

    for reader in range(len(nums)):
        if nums[reader] != val:
            nums[writer] = nums[reader]
            writer += 1

    return nums[:writer]

print(removeElement([3, 2, 2, 3], 3))
print(removeElement([0, 1, 2, 2, 3, 0, 4, 2], 2))
print(removeElement([7, 7, 7], 7))
```

```python
# 12. Find the Missing Number: Find the missing number in an array of size n containing numbers from 1 to n.
# sum of first n natural no = n(n+1)/2 --> XOR solution

def findMissingNumber(nums):
    n = len(nums) + 1

    expected_sum = (n * (n + 1)) // 2

    actual_sum = sum(nums)

    return expected_sum - actual_sum

print(findMissingNumber([1, 2, 4, 5]))
print(findMissingNumber([1]))

'''
def findMissingXOR(nums):
    xor_all = 0
    xor_arr = 0

    # XOR all numbers from 1 to N
    for i in range(1, len(nums) + 2):
        xor_all ^= i

    # XOR all numbers in the array
    for num in nums:
        xor_arr ^= num

    return xor_all ^ xor_arr
'''
```

```python
# 13. Find Duplicates in an Array

def findDuplicatesSet(nums):
    seen = set()
    duplicates = []

    for num in nums:
        if num in seen:
            duplicates.append(num)
        else:
            seen.add(num)

    return duplicates

def findDuplicatesOptimal(nums):
    duplicates = []

    for num in nums:
        index = abs(num) - 1

        if nums[index] < 0:
            duplicates.append(abs(num))
        else:
            nums[index] = -nums[index]

    return duplicates

print(findDuplicatesSet([10, 5, 2, 10, 3, 5]))
print(findDuplicatesOptimal([4, 3, 2, 7, 8, 2, 3, 1]))
print(findDuplicatesOptimal([1, 1, 2]))
```

```python
# 14. Find Intersection of Two Arrays: Find the common elements between two arrays. - O(1)

def intersection(nums1, nums2):
    set1 = set(nums1)
    result = []

    for num in nums2:
        if num in set1:
            result.append(num)
            set1.remove(num)

    return result

print(intersection([1, 2, 2, 1], [2, 2]))
print(intersection([4, 9, 5], [9, 4, 9, 8, 4]))

'''
def intersection(nums1, nums2):
    return list(set(nums1) & set(nums2))
'''
```

```python
# 15. Find Union of Two Arrays

def findUnion(nums1, nums2):
    union_set = set()

    for num in nums1:
        union_set.add(num)

    for num in nums2:
        union_set.add(num)

    return list(union_set)

print(findUnion([1, 2, 3, 4, 5], [1, 2, 3]))
print(findUnion([85, 25, 1, 32, 54, 6], [85, 2]))
print(findUnion([1, 1, 1], [2, 2, 2]))

'''
def findUnion(nums1, nums2):
    return list(set(nums1) | set(nums2))
'''
```

```python
# 16. Check if Two Arrays Are Equal: if two arrays contain the same elements

def checkEqual(nums1, nums2):
    if len(nums1) != len(nums2):
        return False

    freq_map = {}

    for num in nums1:
        if num in freq_map:
            freq_map[num] += 1
        else:
            freq_map[num] = 1

    for num in nums2:
        if num not in freq_map or freq_map[num] == 0:
            return False
        freq_map[num] -= 1

    return True

print(checkEqual([1, 2, 5, 4, 0], [2, 4, 5, 0, 1]))
print(checkEqual([1, 2, 5], [2, 4, 15]))
```

```python
# 17. Find the Leader Elements: An element is a leader if it is greater than all elements to its right.

def findLeaders(nums):
    n = len(nums)
    if n == 0:
        return []

    leaders = []
    max_from_right = nums[n-1]

    leaders.append(max_from_right)

    for i in range(n-2, -1, -1):
        if nums[i] > max_from_right:
            leaders.append(nums[i])
            max_from_right = nums[i]

    return leaders[::-1]

print(findLeaders([16, 17, 4, 3, 5, 2]))
print(findLeaders([1, 2, 3, 4]))
```

```python
# 18. Move Zeroes to End: Move all zeroes in an array to the end while maintaining the order of non-zero elements.

def moveZeroes(nums):
    writer = 0

    for reader in range(len(nums)):
        if nums[reader] != 0:
            nums[writer], nums[reader] = nums[reader], nums[writer]
            writer += 1

    return nums

print(moveZeroes([0, 1, 0, 3, 12]))
print(moveZeroes([0]))
print(moveZeroes([4, 2, 4, 0, 0, 3, 0, 5, 1, 0]))
```

```python
# 19. Find Subarray with Given Sum.

def findSubarraySum(nums, target):
    # Stores {prefix_sum: index}
    prefix_map = {0: -1}
    current_sum = 0

    for i in range(len(nums)):
        current_sum += nums[i]

        # The mathematical check
        if (current_sum - target) in prefix_map:
            start_index = prefix_map[current_sum - target] + 1
            end_index = i

            # Returning the actual elements (could also just return indices)
            return nums[start_index : end_index + 1]

        # Record this sum's first appearance
        if current_sum not in prefix_map:
            prefix_map[current_sum] = i

    return [] # No subarray found

print(findSubarraySum([10, 2, -2, -20, 10], -10))
print(findSubarraySum([1, 4, 20, 3, 10, 5], 33))
print(findSubarraySum([1, 2, 3], 7))
```

```python
# 20. Rotate Array to the Left by k Positions

def rotateLeft(nums, k):
    n = len(nums)
    if n == 0:
        return

    k = k % n

    def reverse(start, end):
        while start < end:
            nums[start], nums[end] = nums[end], nums[start]
            start += 1
            end -= 1

    reverse(0, k - 1)
    reverse(k, n - 1)
    reverse(0, n - 1)

nums1 = [1, 2, 3, 4, 5, 6, 7]
rotateLeft(nums1, 3)
print(nums1)

nums2 = [1, 2, 3, 4, 5]
rotateLeft(nums2, 4)
print(nums2)
```
