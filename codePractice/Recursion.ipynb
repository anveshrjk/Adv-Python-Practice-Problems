# **Recursion**

```python
# sum of array element

def arr_sum(nums):
    if not nums:
        return 0
    return nums[0] + arr_sum(nums[1:])

arr_sum([1,2,3,4,5])
```

```python
# reverse a string

def rev_str(s):
    if len(s) <= 1:
        return s
    return rev_str(s[1:]) + s[0]

rev_str("inte")
```

```python
# factorial

def factorial(n):
    if n == 1 or n == 0:
        return 1
    return factorial(n-1) * n

factorial(3)
factorial(8)
```

```python
# good numbers - leetcode

MOD = 10**9 + 7

def power(x, n, mod):
    if n == 0:
        return 1
    half = power(x, n//2, mod)
    result = (half * half ) % mod
    if n % 2 == 1:
        result = (result * x) % mod
    return result


def count_good_numbers(n):
    even_count = (n+1) // 2
    odd_count = n // 2

    return pow(5, even_count, MOD) * pow(4, odd_count, MOD) % MOD

```
